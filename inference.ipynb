## 혼동 행렬(Confusion Marix)
- 실제 값(정답)과 예측 한 것을 표로 만든 평가표
    - 분류의 예측 결과가 몇개나 맞고 틀렸는지를 확인할 때 사용한다.
- 함수: confusion_matrix(정답, 모델예측값)
- **0번축:** 실제(정답) class, **1번축:** 예측 class, **cell:** 개수
  
![image.png](images/confusion_matrix.png)

### 이진 분류 평가지표
- Accuracy(정확도)
  : 전체 데이터 중에 맞게 예측한 것의 비율, 모든 분류의 기본 평가 방식

#### 양성(Positive) 예측력 측정 평가 지표
- **Recall/Sensitivity (재현율/민감도)**
  - 실제 양성인 것 중에 양성으로 예측한 것의 비율 (TPR)
    - ex) 스팸 메일 중 스팸 메일로 예측한 비율, 금융 사기 데이터 중 사기로 예측한 비율

- **Precision (정밀도)**
  - 양성으로 예측한 것 중에 실제 양성인 비율 (PPV)
    - ex) 스팸 메일로 예측한 것 중 스팸 메일의 비율, 금융 사기 데이터 중 금융 사기인 것의 비율

- **F1 점수**
    - 정밀도와 재현율의 조화 평균 점수
    - recall과 precision이 비슷할수록 높은 값을 가지게 됨.
    - F1 score가 높다는 것은 recall과 precision이 한 쪽으로 치우쳐지지않고 둘 다 좋다고 판단할 수 있는 근거가됨

### 음성(Negative) 예측력 측정 평가 지표
- **Specificity (특이도)**
  - 실제 음성인 것들중 음성으로 맞게 예측한 것의 비율 (TNR)

- **Fall out (위양성률)**
  - 실제 음성인 것들 중 양성으로 잘못 예측한 것의 비율 (FPR)

```python
## DecisionTreeClassifier 검증 ##
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay, 
                            accuracy_score, recall_score, precision_score,
                            f1_score, classification_report)
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier

digits = load_digits()
X = digits.data
y = digits.target
y = np.where(y==9, 1, 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=0)

tree = DecisionTreeClassifier(max_depth=3)

tree.fit(X_train, y_train)

pred_train_tree = tree.predict(X_train)
pred_test_tree = tree.predict(X_test)

cm_train = confusion_matrix(y_train, pred_train_tree)
cm_test = confusion_matrix(y_test, pred_test_tree)

print('train set')
print(cm_train)
print('-----------')
print('test set')
print(cm_test)
```

```python
# 시각화
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

disp_train = ConfusionMatrixDisplay(
    cm_train,
    display_labels=['Not 9', '9']
).plot(cmap="Blues", ax=ax1)

disp_test = ConfusionMatrixDisplay(
    cm_test,
    display_labels=['Not 9', '9']
).plot(cmap='Blues', ax=ax2)
```

```python
# 정확도
print('[Accuracy]')
print(f'Trainset : {accuracy_score(y_train, pred_train_tree)}, Testset:{accuracy_score(y_test, pred_test_tree)}')
print('-------------------------------------------------------------')
print('[Precision]')
print(f'Trainset : {precision_score(y_train, pred_train_tree)}, Testset:{precision_score(y_test, pred_test_tree)}')
print('-------------------------------------------------------------')
print('[Recall]')
print(f'Trainset : {recall_score(y_train, pred_train_tree)}, Testset:{recall_score(y_test, pred_test_tree)}')
print('-------------------------------------------------------------')
print('[f1 score]')
print(f'Trainset : {f1_score(y_train, pred_train_tree)}, Testset:{f1_score(y_test, pred_test_tree)}')
```

```python
## RandomForesstClassifier ##
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=0)

rfc.fit(X_train, y_train)

pred_train_rfc = rfc.predict(X_train)
pred_test_rfc = rfc.predict(X_test)

cm_train_rfc = confusion_matrix(y_train, pred_train_rfc)
cm_train_rfc
```

- 암 환자 판정 모델, 보험 사기 적발 모델


#### 정밀도가 더 중요한 경우
- 실제 음성 데이터를 양성으로 잘못 판단하면 업무상 큰 영향이 있는 경우

- 스팸 메일 판정

- AP Score
  - PR Curve의 성능 평가 지표를 하나의 점수로 평가한 것
  - PR Curve의 선 아래 면적을 계산한 값으로 높을 수록 성능 우수

### ROC curve(Receiver Operating Characteristic Curve)와 AUC(Area Under the Curve) score

- **FPR(False Positive Rate - 위양성율)**
    - 위양성율 (fall-out)
    - 1 - 특이도(TNR)
    - 실제 음성 중 양성으로 잘 못 예측한 비율
    - 낮을 수록 좋음
  $$
    \cfrac{FP}{TN+FP}
  $$

- **TPR(True Positive Rate - 재현율/민감도)**
    - 재현율(recall)
    - 실제 양성 중 양성으로 맞게 예측한 비율
    - 높을 수록 좋음
  $$
    \frac{TP}{FN+TP}
  $$
- Posivite의 임계 값을 변경한 경우 FRP, TPR 비례해서 변화

**ROC Curve**
- 이진 분류의 성능 평가 지표
- FRP 변화할 때 TPR이 어떻게 변화하는지 평가

**AUC Score**
- ROC Curve의 결과를 점수화하는 함수로 ROC Curve 아래쪽 면적 계산
- 0~1  사이의 실수로 나옴 클수록 좋음

![image.png](images/roc.png)
가장 완벽한 것은 FPR이 0이고 TPR이 1인 것이다. 
일반 적으로 FPR이 작을 때 (0에 가까울때) TPR이 높은 경우가 좋은 상황이다. 그래서 선 아래의 면적이 넓은 곡선이 나올 수록 좋은 모델이다.

#### ROC, AUC 점수 확인
- roc_curve(y값, Pos_예측확률) : FPR, TPR, Thresholds
- roc_auc_score(y값, Pos_예측확률) : AUC 점수반환

#### ROC Curve / Precision_Recall Curve
- **ROC Curve/ROC-AUC score**
    - 이진분류에서 양성클래스 탐지와 음성클래스 탐지의 중요도가 비슷할 때 사용(개고양이 분류)
- **Precision Recall Curve/AP Score**
    - 양성클래스 탐지가 음성클래스 탐지의 중요도보다 높을 경우 사용(암환자 진단)

## 회귀(Regression) 평가지표
: 예측할 값이 연속형인 데이터 지도 학습

### 회귀의 주요 평가지표

#### MSE (Mean Squared Error)
- 실제 값과 예측 값의 차를 제곱해 평균 낸 것
- sklearn 평가함수 : mean_squared_error()
- 교차검증 지정 문자열 : neg_mean_squared_error
\begin{align}
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y_i})^2\\
y_i: 실제값, \hat{y_i}: 모델이 예측한 값
\end{align}

#### RMSE(Root Mean Squared Error)
- MSE는 오차를 제곱한 값이므로 실제 오차의 평균보다 큰 값이 나옴 MSE의 제곱근을 계산한 평가지표가 RMSE
- sklearn 평가함수 : root_mean_squared_error()
- 교차검증 지정 문자열 : neg_root_mean_squared_error
    $$
    RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y_i})^2}
    $$

- ### $R^2$ (R square, 결정계수)
  - 결정계수는 회귀모델에서 Feature가 Target을 얼마나 설명한는지 나타내는 평가지표
    - 평균 예측했을 때의 오차보다 모델을 사용했을때 얼마나 더 좋은 성능내는지 비교
    - 1에 가까울수록 좋음
    - sklearn 평가함수 : r2_score()
    - 교차검증 지정 문자열 : r2
$$
    R^2 = \cfrac{\sum_{i=1}^{n}(\hat{y_i}-\bar{y})^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}
    $$
<center>
$y_i$: i번째 정답 값,   
$\hat{y_i}$: i 번째 예측 값,   
$\bar{y}$: y의 평균      
</center>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X,  y = make_regression(
    n_samples=1000,  # 총 데이터 갯수
    n_features=1,     # feature 갯수
    n_informative=1, #  y에 영향을 주는 feature의 갯수
    noise=30,
    random_state=0
)

plt.scatter(X.flatten(), y, alpha=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 모델링
lr = LinearRegression()
tree = DecisionTreeRegressor(max_depth=3, random_state=0)

# 학습
lr.fit(X_train, y_train)
tree.fit(X_train, y_train)

#평가
pred_train_lr = lr.predict(X_train)
pred_test_lr =  lr.predict(X_test)

pred_train_tree = tree.predict(X_train)
pred_test_tree  =  tree.predict(X_test)

# 회귀 평가
print('LinearRegression 평가')
print('MSE : ', mean_squared_error(y_train, pred_train_lr),',',mean_squared_error(y_test, pred_test_lr))
print('RMSE : ', root_mean_squared_error(y_train, pred_train_lr), root_mean_squared_error(y_test, pred_test_lr))
print('R square : ', r2_score(y_train, pred_train_lr), r2_score(y_test, pred_test_lr))

print('--------------------------------------------------')

print("Decision Tree 평가 결과")
print("MSE:", mean_squared_error(y_train, pred_train_tree), mean_squared_error(y_test, pred_test_tree), sep=" , ")
print("RMSE:", root_mean_squared_error(y_train, pred_train_tree), root_mean_squared_error(y_test, pred_test_tree), sep=" , ")
print("R square(결정계수):", r2_score(y_train, pred_train_tree), r2_score(y_test, pred_test_tree), sep=" , ")
```
